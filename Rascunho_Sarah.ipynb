N√£o se esque√ßa de realizar **upload** do Arquivo [Json](https://github.com/SarahFeanor/ada_tech_logica_ii_g2/blob/main/projetoModuloII.json), o programa n√£o funcionar√° sem o arquivo no notebook.

```python
# Importando Bibliotecas necess√°rias
import json

# Op√ß√µes do Menu principal e Submenu do programa:
menu_principal = """
\nBem vindo ao nosso Sistema
\nCadastramento de Usu√°rios:

1 - Inserir usu√°rio
2 - Excluir usu√°rio
3 - Atualizar usu√°rio
4 - Informa√ß√µes de um usu√°rio
5 - Informa√ß√µes de todos os usu√°rios
6 - Sair\n
"""

sub_menu_principal = """
\nQual informa√ß√£o deseja alterar?
1 - Nome
2 - Telefone
3 - Endere√ßo
\n
"""

# Fun√ß√£o 1 - L√™ e salva arquivos:
def ler_salvar_arquivo(ler=True, nome_arquivo='projetoModuloII.json', cadastros=None):
    if ler:
        try:
            with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:  # lembrar de setar o enconding para utf-8
                return json.load(arquivo)
        except FileNotFoundError:
            return {}

    with open(nome_arquivo, 'w') as arquivo:
        json.dump(cadastros, arquivo)


# Fun√ß√£o 2 - Adiciona um novo usu√°rio aos cadastros:
def add_usuario(cadastros, nome, telefone="N√£o Informado", endereco="N√£o Informado"):
    for usuario in cadastros.values():
        if usuario['Nome'] == nome and usuario['Telefone'] == telefone and usuario['Endere√ßo'] == endereco and not usuario['Status']:
            usuario['Status'] = True
            return cadastros

    novo_id = 1
    if cadastros:
        novo_id = max(int(key) for key in cadastros) + 1

    novo_usuario = {
        'Status': True,
        'Nome': nome,
        'Telefone': telefone,
        'Endere√ßo': endereco
    }

    cadastros[str(novo_id)] = novo_usuario
    print(f"Usu√°rio adicionado com sucesso! ID do novo usu√°rio: {novo_id}")
    return cadastros

# Fun√ß√£o 3 - Exclui as informa√ß√µes de um usu√°rio:
def excluir_usuario(cadastros, *ids):
    for id in ids:
        if str(id) in cadastros:
            cadastros[str(id)]['Status'] = False
        else:
            print(f"Usu√°rio com ID {id} n√£o encontrado!")

    ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)
    return cadastros

# Fun√ß√£o 4 - Edita as informa√ß√µes do usu√°rio:
def editar_usuario(cadastros, id, campo, novo_valor):
    if str(id) in cadastros:
        if campo == 1:
            cadastros[str(id)]['Nome'] = novo_valor
        elif campo == 2:
            cadastros[str(id)]['Telefone'] = novo_valor
        elif campo == 3:
            cadastros[str(id)]['Endere√ßo'] = novo_valor
    else:
        print(f"Usu√°rio com ID {id} n√£o encontrado!")

    ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)
    return cadastros

# Fun√ß√£o 5 - Exibe informa√ß√µes do usu√°rio:
def exibir_usuario(cadastros, id):
    if str(id) in cadastros and cadastros[str(id)]['Status']:
        usuario = cadastros[str(id)]
        print(f"Nome: {usuario['Nome']}")
        print(f"Telefone: {usuario['Telefone']}")
        print(f"Endere√ßo: {usuario['Endere√ßo']}")
    else:
        print("Usu√°rio n√£o encontrado!")

# Fun√ß√£o 6 - Exibe informa√ß√µes de todos os usu√°rios ativos:
def exibir_todos_usuarios(cadastros):
    for usuario_id, usuario in cadastros.items():
        if usuario['Status']:
            print(f"ID: {usuario_id}")
            print(f"Nome: {usuario['Nome']}")
            print(f"Telefone: {usuario['Telefone']}")
            print(f"Endere√ßo: {usuario['Endere√ßo']}")
            print()

# Fun√ß√£o Principal que implementa o loop do menu.
# Dependendo da op√ß√£o escolhida, chama as fun√ß√µes apropriadas para realizar opera√ß√µes no cadastro.

def main():
    cadastros = ler_salvar_arquivo()

    while True:
        print(menu_principal)
        opcao = int(input("Informe o n√∫mero da op√ß√£o desejada: "))

        if opcao == 1:
            n = int(input("Informe o n√∫mero de cadastros que deseja inserir: "))
            for _ in range(n):
                nome = input("Insira o nome: ")
                telefone = input("Insira o telefone: ")
                endereco = input("Insira o endere√ßo: ")
                cadastros = add_usuario(cadastros, nome, telefone, endereco)
            ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)

        elif opcao == 2:
            ids = [int(id) for id in input("Insira o(s) ID(s) do(s) usu√°rio(s) que deseja excluir (separados por espa√ßo): ").split()]
            cadastros = excluir_usuario(cadastros, *ids)
            ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)

        elif opcao == 3:
            id_usuario = int(input("Insira o ID do usu√°rio que deseja atualizar: "))
            print(sub_menu_principal)
            campo = int(input("Digite a op√ß√£o desejada: "))
            novo_valor = input(f"Insira o novo valor para o campo selecionado: ")
            cadastros = editar_usuario(cadastros, id_usuario, campo, novo_valor)
            ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)

        elif opcao == 4:
            id_usuario = int(input("Insira o ID do usu√°rio que deseja exibir: "))
            exibir_usuario(cadastros, id_usuario)

        elif opcao == 5:
            exibir_todos_usuarios(cadastros)

        elif opcao == 6:
            ler_salvar_arquivo(ler=False, nome_arquivo='projetoModuloII.json', cadastros=cadastros)
            print("Programa encerrado.")
            break

        else:
            print("Op√ß√£o inv√°lida!")

if __name__ == "__main__":
  main()
```

# üíª Explica√ß√£o do c√≥digo



- Esse c√≥digo em Python √© um programa de gerenciamento de cadastros de usu√°rios que utiliza um arquivo JSON para armazenar as informa√ß√µes dos usu√°rios.

- Oferece funcionalidades b√°sicas de gerenciamento de cadastros, como adi√ß√£o, exclus√£o, edi√ß√£o e visualiza√ß√£o de usu√°rios. Os dados s√£o armazenados e recuperados de um arquivo JSON para persist√™ncia entre diferentes execu√ß√µes do programa.

### üî∂ Explica√ß√£o de cada parte do c√≥digo:

1. **Menu e Submenu:**
   - `menu_str` e `sub_menu_str` s√£o strings que representam o menu principal e o submenu do programa. Eles s√£o utilizados para exibir op√ß√µes ao usu√°rio.

2. **Fun√ß√£o `ler_salvar_arquivo`:**
   - Esta fun√ß√£o l√™ ou salva informa√ß√µes em um arquivo JSON.
   - Se `ler` for `True`, ela tenta abrir o arquivo para leitura e carrega os dados do arquivo JSON usando `json.load()`. Se o arquivo n√£o existir, retorna um dicion√°rio vazio.
   - Se `ler` for `False`, ela abre o arquivo para escrita (`'w'`) e salva os dados usando `json.dump()`.

3. **Fun√ß√£o `add_usuario`:**
   - Adiciona um novo usu√°rio aos cadastros. Se um usu√°rio com o mesmo nome, telefone e endere√ßo j√° existir e estiver inativo, ativa esse usu√°rio. Caso contr√°rio, cria um novo usu√°rio.

4. **Fun√ß√£o `excluir_usuario`:**
   - Marca usu√°rios como inativos com base nos IDs fornecidos.

5. **Fun√ß√£o `editar_usuario`:**
   - Edita as informa√ß√µes de um usu√°rio com base no ID, campo (nome, telefone, endere√ßo) e novo valor fornecidos.

6. **Fun√ß√£o `exibir_usuario`:**
   - Exibe informa√ß√µes de um usu√°rio com base no ID fornecido.

7. **Fun√ß√£o `exibir_todos_usuarios`:**
   - Exibe informa√ß√µes de todos os usu√°rios ativos.

8. **Fun√ß√£o `main`:**
   - Fun√ß√£o principal que implementa o loop do menu. O programa continua executando at√© que o usu√°rio escolha a op√ß√£o 6 para sair.
   - Dependendo da op√ß√£o escolhida, chama as fun√ß√µes apropriadas para realizar opera√ß√µes no cadastro.

9. **Bloco `if __name__ == "__main__":`**
   - Garante que o programa seja executado apenas quando o script √© executado diretamente, n√£o quando √© importado como um m√≥dulo.
